## Problem Statement
- We need to analyze the youth global unemployment dataset for last five years , from2017 to 2021 to get the answers for the following points and plot the results :
    1. List all the countries where youth unemployment is greater than or equal to 25% in 2021.
    2. List all countries where youth unemployment is less than or equal to 1% in2021
    3. Top 10 countries that have an average high rate of youth unemployment from2017 to 2021
    4. Top 10 countries that have an average low rate of youth unemployment from2017 to 2021
    5. Top 10 countries that have a high COVID impact (for years 2019 and 2020)
    6. Top 10 countries that have a less COVID impact (for years 2019 and 2020)


## Step by Step to perform data analysis

### 1. Importing The Dataset

```python
df=pd.read_csv('unemp.csv')
```

### 2. Exploring,Selecting,Cleaning and preparing the data

```python
df.columns
```

```python
df.head()
```

```python
df.shape
```

```python
# Creating a new dataframe with 5 years data
df5=df.iloc[0:,[0,1,28,29,30,31,32]]
```

```python
# Check for nulls
df5.isnull().sum()
```

```python
df5.dropna()
```

```python
# checking for duplicated
df5.duplicated().sum()
```

```python
df5.dtypes
```

```python
# sorting the DataFrame
df5.sort_values('Country Name')
```

```python
# setting the index(setting COUNTRY CODE column as an index of the following coding snippet)
df5.set_index('Country Code',inplace=True)
```

```python
df5
```

```python
# Adding two new columns 5yrsavg and covid year difference
df5['5yrsavg']=df5[['2017','2018','2019','2020','2021']].mean(axis=1)
df5['covid_yr_diff']=df5['2020']-df5['2019']
```

```python
df5
```

## plotting

```python
# plotting the 5 year average unemployment data
```

```python
avg5yrs=df5['5yrsavg']# x values
countries=df5.index
plt.figure(figsize=(12,6))
plt.bar(countries,avg5yrs)
plt.xlabel('Countries')
plt.ylabel('5-Year Average')
plt.title('5-Year Average by Country')
plt.tight_layout()
plt.show()
```

### 1. Performing the data analysis and plotting the summary of them

###### i.List all countries where youth unemployment is greater than or equal to25% in 2021.

```python
#List all countries where youth unemployment is greater than or equal to 25% in 2021.
df5[df5['2021']>25][['Country Name','2021']]

```

###### ii. List all countries where youth unemployment is less than or equal to 1% in2021.

```python
#List all countries where youth unemployment is less than or equal to 1% in2021.
df5[df5['2021']<=1][['Country Name','2021']]
```

###### iii. Top 10 countries that have an average high rate of youth unemployment from 2017 to 2021.

```python
top10cont=df5.sort_values('5yrsavg',ascending=False)[0:10]
```

```python
top10cont
```

```python
# plot of the top 10 countries with highest unemployment rate
```

```python
uemp=top10cont['5yrsavg']
```

```python
countries=top10cont.index
```

```python
plt.bar(countries,uemp)
plt.xlabel('countrycode',fontsize=12)
plt.ylabel("5 Yrs Avg Unemployment Rate",fontsize=14)
plt.xticks(rotation="vertical",fontsize=12)
plt.grid()
plt.show()
```

######  iv.Top 10 countries that have an average low rate of youth unemployment from 2017 to 2021

```python
top10cont_lowunemp=df5.sort_values('5yrsavg')[0:10]
```

```python
# plotting
lowunemp=top10cont_lowunemp['5yrsavg']
countries=top10cont_lowunemp.index
plt.bar(countries,lowunemp)
plt.xlabel('countrycode',fontsize=12)
plt.ylabel('Top 10 countries with low unemployment rate',fontsize=12)
plt.show()
```

##### v. Top 10 countries that have a high covid impact on the unemployment rate (for the immediate year 2019 and covid year 2020)

```python
df5
```

```python
highcovidimpact=df5.sort_values('covid_yr_diff',ascending=False)[0:10]
highcovidimpact
```

```python
coviddiffdata=highcovidimpact['covid_yr_diff']
```

```python
# plotting
plt.bar(highcovidimpact.index,coviddiffdata)
plt.xlabel('Country Code',fontsize=12)
plt.ylabel("Rate")
plt.grid()
plt.show()
```

###### vi. Top 10 countries that have a less covid impact (for the immediate year 2019 and the covid year 2020)

```python
lowimpact=df5.sort_values('covid_yr_diff')[0:10]
lowimpact
```

```python
# Plotting
datas=lowimpact['covid_yr_diff']

plt.bar(lowimpact.index,datas)
plt.xticks(rotation=45)
plt.show()
```
